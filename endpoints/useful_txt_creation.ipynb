Lo que voy a hacer en este archivo es crear un archivo en formato txt que contenga todos los valores posibles para ingresar a los correspondientes endpoints. Por ejemplo, para el endpoint 1 que hay que ingresar un desarrollador, tendría la lista completa de todos los desarrolladores posibles. 
Esto lo hago para que la persona que use los endpoint tenga fácil acceso a los datos y pueda ver los resultados de forma clara.

Empecemos con el primer endpoint

```python
# Primero importamos las librerías y los archivos necesarios para trabajar
import pandas as pd
import csv
import sys
dev_prices_year = pd.read_parquet("../Datasets/endpoint_1/Steam_games_endpoint_1.parquet")
dev_prices_year.head()
```

```python
# Lo que nos improta acá es la columna de developer, ya que eso se ingresa a la función, así que:
lista_desarrolladores = list(dev_prices_year["developer"].unique())
lista_desarrolladores
```

```python
# Ahora, esta lista la voy a almacenar cómo valor dentro de un diccionario, la clave va a ser el nombre de la función:
parametros_validos = {}
parametros_validos['Desarrolladoras'] = lista_desarrolladores
parametros_validos
```

Ahora voy con el segundo endpoint.

```python
user_items = pd.read_parquet("../Datasets/endpoint_2/users_items.parquet")
user_items.head()
```

```python
lista_usuarios = list(user_items["user_id"].unique())
lista_usuarios
```

```python
# Nuevamente, relleno el diccionario con los valores correspondientes a la lista
parametros_validos["Usuarios"] = lista_usuarios
parametros_validos
```

Tercer endpoint

```python
juegos = pd.read_parquet("../Datasets/steam_games_complete.parquet")
juegos
```

```python
juegos = juegos.explode("genres")
juegos.head()
```

```python
lista_generos = list(juegos["genres"].unique())
lista_generos
```

```python
# Nuevamente, agrego esta lista al diccionario
parametros_validos["Generos"] = lista_generos
```

Cuarto endpoint

```python
desarrolladores_y_reseñas = pd.read_parquet("../Datasets/endpoint_4/desarrolladores_y_reseñas.parquet")
desarrolladores_y_reseñas.head()
```

```python
lista_años = list(desarrolladores_y_reseñas["año"].unique())
lista_años
```

```python
parametros_validos["Años"] = lista_años
```

Quinto endpoint

```python
# Para este endpoint también hay que ingresar a la función un desarrollador, cómo ya lo había hecho en el primer endpoint, no lo repito
# Escribamos ahora el diccionario en un archivo .txt
with open("parametros_validos_endpoints.txt", "w", encoding="utf-8") as file:
    for funcion, parametros in parametros_validos.items(): # Por cada función (clave) y sus parametros (valor / lista) 
        file.write(f"{funcion}:\n") # Escribir la función de cabecera
        for parametro in parametros: # Luego iterar sobre la lista
            file.write(f"  - {parametro}\n") # Y darle un formato lista
        file.write("\n")  
```
