Al igual que lo hice en la carpeta de los endpoints, en este archivo voy a crear un archivo .txt con los parametros válidos para ambos sistemas de recomendación.

Empecemos con el primer sistema de recomendación (item-item)

```python
# Importamos las librerías necesarias para trabajar
import pandas as pd
```

```python
# Cargamos el archivo de juegos, el cuál contiene en una columna los ID de los juegos, los cuáles son los parametros válidos de la función
juegos = pd.read_parquet("../Datasets/recomendacion/uno/steam_games_id_name_categorical.parquet")
juegos
```

Almacenamos los parametros válidos en una lista

```python
lista_de_id_juegos = list(juegos["item_id"].unique())
lista_de_id_juegos
```

Ahora, esta lista la almaceno en un diccionario cómo valor, que cómo clave va a tener "IDs válidos (recomendación_juego)"

```python
parametros_validos = {}
parametros_validos["IDs válidos (recomendacion_juego)"] = lista_de_id_juegos
parametros_validos
```

Ahora voy con el segundo sistema de recomendación

```python
# Los parametros válidos que tiene esta función son los usuarios, los mismos se encuentran en la tabla similitud entre usuarios
similitud = pd.read_parquet("../Datasets/recomendacion/dos/similitud_entre_usuarios.parquet")
similitud
```

```python
lista_de_id_usuarios = list(similitud.columns)
lista_de_id_usuarios
```

Ahora lo agrego al diccionario

```python
parametros_validos["IDs válidos (recomendacion_usuario)"] = lista_de_id_usuarios
parametros_validos
```

Ahora a partir de este diccionario, creo un archivo .txt que contenga los paramatros válidos

```python
# Escribamos ahora el diccionario en un archivo .txt
with open("parametros_validos_sistema_recomendacion.txt", "w", encoding="utf-8") as file:
    for funcion, parametros in parametros_validos.items(): # Por cada función (clave) y sus parametros (valor / lista) 
        file.write(f"{funcion}:\n") # Escribir la función de cabecera
        for parametro in parametros: # Luego iterar sobre la lista
            file.write(f"  - {parametro}\n") # Y darle un formato lista
        file.write("\n")  
```
